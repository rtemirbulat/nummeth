```python
import numpy as np 
import matplotlib.pyplot as plt 
```

### Half-division and Secant method

```python
#prereq
a = 0
b = 2
iterations = 100
x = np.linspace(-2,2,100)
f = lambda x: x**4 -x - 10
y = x**4 - x -10
```

```python
#plot
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x,y, 'g')
ax.grid()
plt.show()
```

```python
def half_div(f,a,b,iterations):
    if (f(a)*f(b)) >= 0:
        print("Method fails.")
        return None
    a_n = a
    b_n = b
    for n in range(1,iterations+1):
        c_n = (a_n + b_n)/2 #distance
        f_c_n = f(c_n)
        if f(a_n)*f_c_n < 0:
            a_n = a_n
            b_n = c_n
        elif f(b_n)*f_c_n < 0:
            a_n = c_n
            b_n = b_n
        elif f_c_n == 0:
            print("Found exact solution.")
            return c_n
        else:
            print("Bisection method fails.")
            return None
    return (a_n + b_n)/2
```

```python
def secant_method(f,a,b,iterations):
    if f(a)*f(b) >= 0:
        print("Secant method fails.")
        return None
    a_n = a
    b_n = b
    for n in range(1,iterations+1):
        c_n = a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
        f_c_n = f(c_n)
        if f(a_n)*f_c_n < 0:
            a_n = a_n
            b_n = c_n
        elif f(b_n)*f_c_n < 0:
            a_n = c_n
            b_n = b_n
        elif f_c_n == 0:
            print("Found exact solution.")
            return c_n
        else:
            print("Secant method fails.")
            return None
    return a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
```

```python
print(half_div(f,a,b,iterations))
print(secant_method(f,a,b,iterations))
```
